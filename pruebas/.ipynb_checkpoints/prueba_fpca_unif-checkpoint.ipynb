```python
import sys
sys.path.append('../')
from utils import *
SEED = 0
```

```python
# FPCA
N_BASIS = 30
N_COMPONENTS = 2
```

```python
# Para generaci√≥n de datos
VARS = ['var1']
T = 20
H_STEPS = 5
TIME_PERIODS = np.arange(T)
PREDICT_RANGE = range(T, T+H_STEPS)
N_SAMPLES = 1000
```

```python
#np.random.seed(SEED)
#torch.manual_seed(SEED)

sigma = 1 # 0.3
offset = 10
```

```python
data = []

for t in TIME_PERIODS:
    samples = np.random.normal(loc=t, scale=sigma, size=[N_SAMPLES,len(VARS)])
    samples = pd.DataFrame(samples, columns = VARS)
    samples['time'] = t
    data.append(samples)

data = pd.concat(data, ignore_index=True)

# Ajusto KDEs
kdes = get_kdes(data, VARS[0], 'time', TIME_PERIODS)

grid_points = np.linspace(data.var1.min()-offset, data.var1.max()+offset, N_SAMPLES)
data_matrix = [kdes[t].score_samples(grid_points.reshape(-1,1)) for t in TIME_PERIODS]

# Genero hasta H_STEPS
true = []

for t in PREDICT_RANGE:
    samples = np.random.normal(loc=t, scale=sigma, size=[N_SAMPLES, len(VARS)])
    samples = pd.DataFrame(samples, columns=VARS)
    samples['time'] = t
    true.append(samples)

true = pd.concat(true, ignore_index=True)

true_kdes = get_kdes(true, VARS[0], 'time', PREDICT_RANGE)

true_data_matrix = [true_kdes[t].score_samples(grid_points.reshape(-1,1)) for t in PREDICT_RANGE]
```

```python
for i, d in zip(np.linspace(0.3,0.7,T), data_matrix):
    plt.plot(grid_points, np.exp(d), color=BLUE(i))
plt.gca().set_ylim(bottom=0);
```

```python
fpca_unif = Fpca(data_matrix, grid_points, n_basis=N_BASIS)
fpca_unif.fit_fpca(n_components=N_COMPONENTS)

models = {}

models[0] = FitData('lin')
models[0].fit(TIME_PERIODS, fpca_unif.scores[:,0])

models[1] = FitData('poly', degree=3)
models[1].fit(TIME_PERIODS, fpca_unif.scores[:,1])

# models[2] = FitData('poly', degree=3)
# models[2].fit(TIME_PERIODS, fpca_unif.scores[:,2])

# models[2] = auto_arima(fpca_unif.scores[:,2], start_p=0, d=3, start_q=0,
#                      max_p=10, max_d=10, max_q=10, start_P=0,
#                      D=1, start_Q=0, max_P=10, max_D=10,
#                      max_Q=10, m=1, seasonal=False,
#                      error_action='warn', trace=False,
#                      stepwise=True, random_state=20, n_fits=1000 )

# for i in range(N_COMPONENTS):
#     models[i] = auto_arima(fpca_unif.scores[:,i], start_p=0, d=2, start_q=0,
#                              max_p=10, max_d=10, max_q=10, start_P=0,
#                              D=1, start_Q=0, max_P=10, max_D=10,
#                              max_Q=10, m=1, seasonal=False,
#                              error_action='warn', trace=False,
#                              stepwise=True, random_state=20, n_fits=1000 )

#models[2] = FitData('poly', degree=3)
#models[2].fit(TIME_PERIODS, fpca_unif.scores[:,2])

fpca_unif.fit_time_series(models=models, h_steps=H_STEPS)

# for i in range(1, N_COMPONENTS):
#     fpca_unif.pred_w[:,i] = [fpca_unif.scores[-1,i]]*H_STEPS # dejamos este score constante

fpca_unif.predict(exp=True, normalize=True)

kls_f = [get_kl( fpca_unif.pred_f[i][1:-1], np.exp(true_data_matrix[i])[1:-1] ) for i in range(H_STEPS)]
kl_f = np.mean(kls_f)
```
